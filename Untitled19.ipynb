```python
contacts={}
while(True):
    print("1:Add new contact")
    print("2:Search on someone")
    print("3:Exit")
    choise=int(input("Enter Choise: "))

    if(choise==1):
        name=input("Enter the name: ")
        number=int(input("Enter the number: "))
        contacts[name]=number
        print("New contact was added")
        print("***********************************************************")

    elif(choise==3):
        print("Exit")
        break
    elif(choise==2):
        name=input("Enter the name: ")
        if(name in contacts):
            print(f"the name is{name}")
        else:
            print("not exict")

            print("***********************************************************")
```
